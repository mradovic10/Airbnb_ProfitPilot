```python
import pandas as pd
import hvplot.pandas
import numpy as np

```

```python
df3=pd.read_csv("../Resources/cleaned_data_SD_final.csv", dtype={"bathrooms_cat": np.object,
                                                                "bedrooms_cat": np.object,
                                                                "zipcode": np.object,
                                                                "years_in_business_cat": np.object})
df3=df3.iloc[: , 1:]
df3.head()
```

```python
df3.info()
```

```python
df3["zipcode"]=df3["zipcode"].astype(object)
```

```python
df3["bathrooms_cat"]=df3["bathrooms_cat"].astype(object)
```

```python

df3["bathrooms_cat"].unique()
```

```python
df3["bedrooms_cat"].unique()
```

```python
df3.info()
```

```python
# find the average price and standrad error of the mean per bedroom
avg_price=pd.DataFrame(df3.groupby("bedrooms_cat")["price"].agg(['mean', 'sem', 'min', 'max', 'count']))
avg_price["LCI"]=avg_price['mean'] - 1.96* avg_price['sem']
avg_price["UCI"]=avg_price['mean'] + 1.96* avg_price['sem']
avg_price["error1"]=avg_price['mean'] - avg_price['min']
avg_price["error2"]=avg_price['max'] - avg_price['mean']
avg_price
```

```python
Fig1=avg_price.hvplot.bar(x="bedrooms_cat", y="mean", alpha=0.6, color="lime")
Fig2=avg_price.hvplot.errorbars(x='bedrooms_cat', y='mean', yerr1='error1',yerr2='error2', color='black')
Fig3=avg_price.hvplot.labels(x='bedrooms_cat', y= 'mean', text='count')

Fig12= Fig1 * Fig2 
Fig4=Fig12*Fig3

Fig4=Fig4.options(xlabel='Bedrooms',
                  ylabel='Average Price ($)',
                  fontsize={'title': 14,'labels': 14,'xticks': 12,'yticks': 12},
                  title='Average price with minimum and maximum price shown as error bars')
hvplot.save(Fig4, 'Average_price.png', fmt='png')
```
